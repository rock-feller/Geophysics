```python
%pylab inline
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from matplotlib.mlab import griddata



def dame_xyz(filename):
    file = open(filename, 'r') #Open files from Brahe-3141-f/ directory
    x, y, z= loadtxt(file, unpack=True)  # http://bulldog2.redlands.edu/facultyfolder/deweerd/tutorials/ReadingWriting.html
    size_of_x = size(y) #gives how many lines there are in the file
    return x,y,z
    

def dame_Z1XY(filename):
    x,y,z = dame_xyz(filename)
    max_a = int(max(x)/5)
    max_b = int(max(y)/5)
    Z1 = ones((max_b+1, max_a+1))
    size_of_x = size(x)
    X = []
    Y = []
    
    for k in range(max_a + 1):
        X.append(k*5)
    for h in range(max_b +1):
        Y.append(h*5)

    for i in range(size_of_x):
        a = x[i]/5
        b = y[i]/5
        if(Z1[b][a] == 1):
            Z1[b][a] = z[i]
        if(Z1[b][a] != 0):
            Z1[b][a] = ( Z1[b][a] + z[i] )/ 2    
        
    return Z1,X,Y   
    
    
#for i in range(int(max_b)):
 #   for j in range(int(max_a)):
  #      Z[i][j] = 0
def plot_survey(filename):
    x,y,z = dame_xyz(filename)
    plt.plot(x,y)
    plt.title('Survey 1 - Grupo A')
    plt.ylabel('Distancia Recorrida en Y')
    plt.xlabel('Distancia Recorrida en X')
    plot(0,0,'ro')
    plt.savefig(filename +'.png')
    plt.show()

#Plot2
    
def plot2_survey(filename):
    
    x,y,z = dame_xyz(filename)
    Z1,X,Y = dame_Z1XY(filename)
    plot_survey(filename)
        
    fig = plt.figure()
    fig.set_size_inches(10.5,5.5)
    ax = fig.gca(projection='3d')

    xi = np.linspace(min(x),max(x))
    yi = np.linspace(min(y),max(y))

    X1,Y1 = meshgrid(X,Y)
    X, Y = np.meshgrid(xi, yi)
    Z = griddata(x, y, z, xi, yi)

    surf = ax.plot_surface(X, Y, Z)
    ax.scatter(X1,Y1,Z1,zdir=u'z', s=20, c=u'r')
    ax.set_zlim3d(np.min(Z), np.max(Z))

    ax.set_xlabel('Distancia en x',fontsize=15)
    ax.set_ylabel('Distancia en y',fontsize=15)
    ax.set_zlabel('Intensidad de Campo Magnetico medida (nT)',fontsize=20)
    ax.set_title('Campo Magnetico',fontsize=25)
    plt.savefig('Intensidad_Mag_'+filename+'.png')
```
